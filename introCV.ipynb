```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# 載入影像

影像也是資料的一種。

`cv2.imread`就可以將影像資料轉換成我們熟悉的資料型別 - Numpy array

```python
# 載入影像
image = cv2.imread("sim_data/images/plane.jpg")

# 顯示影像
plt.imshow(image, cmap="gray"), plt.axis("off")
plt.show()
```

```python
# 載入影像成灰階影像
image = cv2.imread("sim_data/images/plane.jpg", cv2.IMREAD_GRAYSCALE)

# 顯示影像
plt.imshow(image, cmap="gray"), plt.axis("off")
plt.show()
```

影像矩陣中的每一元素都分別對應到影像上的每一個像素（pixels）

```python
# 影像矩陣
image
```

此影像的解析度是 3600 * 2270，所以該矩陣的尺寸/維度也會是這樣

```python
# 顯示該影像的矩陣維度
image.shape
```

那矩陣中每一個元素所代表的是什麼？

就灰階影像而言，每一個元素的值，就是對應像素的亮度（intensity）。

亮度值能從黑（0）到白（255）。

```python
# slicing the very first element in array
image[0, 0]
```

假設讀入的影像是彩色的，在它所構成的矩陣中，每一個元素包含有三個值，分別對應藍、綠和紅（BGR）的值。

```python
# 以彩色方式載入影像
image_bgr = cv2.imread("sim_data/images/plane.jpg", cv2.IMREAD_COLOR)

# 顯示影像
plt.imshow(image_bgr), plt.axis("off")
```

```python
# slicing the very first element in array
image_bgr[0, 0]
```

注意：`OpenCV`預設使用`BGR`，但許多影像軟體則使用紅、綠和藍（`RBG`）。

所以想在 matplotlib 中正確顯示 openCV 的彩色影像，需要先將顏色轉成 RGB。

```python
# 轉成 RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# 顯示影像
plt.imshow(image_rgb), plt.axis("off")
plt.show()
```
